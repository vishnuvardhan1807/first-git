<font size="5">now lets see of iterating through rows and columns

```python
import pandas as pd
df = pd.read_csv("D:\\data science\\pandas\\csv\\nba.csv")
c=df.head(10)
c

count = 0

for i,j in c.iterrows():
    count = count + 1
    if count<4:
        print(i,j)
        print()

```

```python
for i,j in c.iteritems():
    print(i,j)
    print()
```

```python
count = 0
for i in c.itertuples():
    print(i)
    print()
```
