<font size="5"><font color="red">inorder to get Nan to specied values we have to pas stwo values to replace parameter one with the value we want to change and another with np.NaN

```python
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
new_df = df.replace([32,32],np.NaN)
new_df
```

<font size="5"><font color="blue">if we want to change vaues particularly in different rows of different coulumns then we use dictionaries with col names

```python
new_df = df.replace({'temparature':-99999,
                    'event':"no Event"},np.NaN)
new_df
```

<font size="5">we can even use mappings through dictionaries to achieve our goal 

```python
new_df = df.replace({-99999:np.NaN,
                    'no Event':np.NaN})
new_df
```

<font size="5"><font color="pink>we can also change by using lists to send data in replace function

```python
new_df = df.replace([32,31,6,2],["ra","po","mg","ou"])
new_df
```
